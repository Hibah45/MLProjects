```python
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
x["area_mean"].describe()
```

```python
from sklearn import cross_validation, neighbors #knn nearest neighbors
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test=y_test.astype('int')
y_test
```

```python
y_train=y_train.astype('int')
y_train
```

# kNN Classifier

```python
#Working wid knn classifier-Fitting Classifier to training set
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
#predict on test data
y_predict=clf.predict(x_test)
```

```python
print("Actual breast cancer:")
print(y_test.values)
```

```python
print("Predicted cancer values:")
print(y_predict)
```

```python
#Accuracy score 
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix

print("\n Accuracy score:%f" %(accuracy_score(y_test,y_predict)*100))
print("\n Recall score:%f" %(recall_score(y_test,y_predict)*100))
print("\n ROC score:%f\n" %(roc_auc_score(y_test,y_predict)*100))
print(confusion_matrix(y_test,y_predict))
```

```python
#Roc- Receiver operating characteristics curve
#good model-more TP,tolerable FP,min False negatives
```

```python
#predict the probability
probas=clf.predict_proba(x_test)
probas
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=150)
plt.hist(probas,bins=20)
plt.title("Classification Probabilities")
plt.xlabel("Probability")
plt.ylabel('# of Instances')
plt.xlim([0.5,1.0])
plt.legend(y_test) #legends r classifiers-color
plt.show()
```

```python
x_train_std=minmax.fit_transform(x_train)
x_test_std=minmax.transform(x_test)
```

```python
#Evaluating metrics
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
clf_acc=cross_val_score(clf,x_train_std,y_train,cv=3,scoring="accuracy",n_jobs=-1) #cv-cross validation
clf_acc
```

```python
clf_proba=cross_val_predict(clf,x_train_std,y_train,cv=3,method='predict_proba') #cv-cross validation
clf_proba
```

```python
clf_scores=clf_proba[:,1] #get score of positive class
clf_scores
```

# Logistic Regression

```python
#build a logistic Regression model-used when output var is categorical
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)

lr_acc=cross_val_score(lr,x_train_std,y_train,cv=3,scoring='accuracy',n_jobs=-1)
lr_proba=cross_val_predict(lr,x_train_std,y_train,cv=3,method='predict_proba')

lr_scores=lr_proba[:,1]
```
